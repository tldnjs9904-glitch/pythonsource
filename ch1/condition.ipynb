````python
# 한 시간 임금은 10700원, 8시간 까지는 10700 으로 지급하고 8시간 초과 근무시에는 1.5 배 지급

# 근무시간(hours), rate(시간당 임금), pay(일당)
hours,rate,pay = 12, 10700, 0

if hours <= 8:
    pay = hours * rate
else:
    pay = (8 * rate) + ((hours - 8) * rate * 1.5)

print(f"근무시간: {hours}시간")
print(f"일당: {pay:,.0f}원")
````

````python
# 주차장 프로그램
# 사용자에게 메뉴 제공
# [1] 주차 | [2] 출차 | [3] 종료

# 사용자는 메뉴 선택
# 1
# A 자동차 입차. 주차장 상태 ==> ['A']
# 1
# B 자동차 입차. 주차장 상태 ==> ['A'], ['B']
# 2
# B 자동차 출차. 주차장 상태 ==> ['A']
# E 자동차 입차. 주차장 상태 ==> ['A', 'B', 'C', 'D', 'E']
# 1
# 주차장 만차

parking_lot = []
top, car_name = 0, 'A'

while True:
    no = int(input("[1] 주차 | [2] 출차 | [3] 종료"))

    if no <= 3:
        if no == 1:
            # top >= 5 주차장 만차 => 아무것도 안함
            if top >= 5:
                print("주차장 만차")
            else:
                parking_lot.append(car_name)
                # 출력 : A 자동차 입차. 주차장 상태 ==> ['A']
                print(f"{car_name} 자동차 입차. 주차장 상태 ==> {parking_lot}")
                car_name = chr(ord(car_name)+1)
                top = top +1
        elif no == 2:
            if top > 0:
                # 리스트 뒤부터 제거 : pop()
                outCar= parking_lot,pop()
                print(f"{outCar}자동차 출차. 주차장 상태 ==> {parking_lot}")
                top = top-1
                car_name = chr(ord(car_name)-1)
            else:
                print("빠져나갈 자동차가 없음")
        else:
            print("프로그램 종료")
            break
    else:
        print("번호를 확인해 주세요")
````

````python
parking = []
max_parking = 5

while True:
    print("\n[1] 주차 | [2] 출차 | [3] 종료")
    menu = input("메뉴를 선택하세요: ")

    # 주차
    if menu == "1":
        if len(parking) >= max_parking:
            print("주차장 만차")
            continue

        car = input("자동차 이름: ")
        parking.append(car)

        print(f"{car} 자동차 입차")
        print(f"주차장 상태 ==> {parking}")

    # 출차
    elif menu == "2":
        if len(parking) == 0:
            print("주차장이 비어 있습니다.")
            continue

        car = parking.pop()

        print(f"{car} 자동차 출차")
        print(f"주차장 상태 ==> {parking}")

    # 종료
    elif menu == "3":
        print("주차장 프로그램을 종료합니다.")
        break

    # 잘못된 메뉴
    else:
        print("1, 2, 3 중 하나를 선택하세요.")
````
