````python
# add2
# add2(2,3)
# add2(2,3,4,5)
# add2(2,3,4,5,6,7,8)

# 가변 매개변수
# 입력값이 몇 개가 될 지 모르는 경우
# *
def add2(*args):
    result = 0
    for i in args:
        result = result + i
    print(result)

````

````python
# def add_mul(choice, *args, a):
#     pass

# 일반 매개변수는 앞 쪽으로 선언
# +, * 실행
def add_mul(choice, *args):
    if choice == "add":
        result = 0
        for i in args:
            result = result + i


    elif choice == "mul":
        result = 1
        for i in args:
            result = result * i

    return result

````

````python
# add_mul("add",10,11,12,13,a=10)
print(add_mul("add",10,11,12,13))
print(add_mul("mul",10,11,12,13))
````

````python
# 키워드 매개변수 (key=value)
# **변수명

# 딕셔너리 == 키워드 매개변수
def print_kwargs(**kwargs):
    print(kwargs)

    # key값만 출력
    for k in kwargs.keys():
        print(k)
    

    # value 값만 출력
    for k,v in kwargs.items():
        print(k,v)
````

````python
print_kwargs(name="lee", age=15)
print_kwargs(name="si",age=25, addr="서울")
print_kwargs(name="won",age=28, addr="서울특별시", tel="010-5818-5818")
````

````python
#  변수의 효력 범위

# 함수 밖의 변수 : 전역변수
a = 1
def vartest(a):
    # 매개변수 a : 지역변수
    a = a + 1
    return a

a = vartest(a)
print(a)
````

````python
# 전역변수 a 값 변경하기.
# 1) 리턴 이용(추천방법) 추천이유 : 안정적인 방법
a = 1 
def vartest(a):
    # 매개변수 a : 지역변수
    a = a + 1
    return a

a = vartest(a)
print(a)
````

````python
# 2) global 키워드 사용
# 함수 안에서 전역변수 접근 안됨

a = 1 
def vartest(): 
    global a
    a = a + 1
    

vartest()
print(a)
````

````python
# 사칙연산 함수
# 결과 리턴
def calculator(num1,num2,op):
    # if~elif
    if op == "+":
        return num1 + num2
    elif op == "-":
        return num1 - num2
    elif op == "*":
        return num1 * num2
    elif op == "/":
        return num1 / num2
    elif op == "//":
        return num1 // num2

print(calculator(10,20,"+"))
print(calculator(10,20,"-"))
print(calculator(10,20,"*"))
print(calculator(10,20,"/"))
print(calculator(10,20,"//"))
````
