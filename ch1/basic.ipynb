#### 문자열 포매팅
- 문자열 안에 어떤 값을 삽입하는 방법

| 코드 | 설명 |
| --- | --- |
|`%s`| 문자열(string)|
|`%d |정수 (integer)   |
| %c | 문자 1개(character) |
| %f | 부동 소수 (floating-point) |
| %o | 8진수 |
| %x | 16진 |
%%| % 문자 자체|

- format() 함수 사용
- f 문자열 포매팅(파이썬3.6버전부터 지원)


```python
# 학습 포인트: %d 자리에 정수 값을 넣는 전통적인 문자열 포매팅 방식입니다.
# %s => 어떤 값이든 대입 가능
a = "i eat %d apples" % 10
```

```python
# 학습 포인트: 문자열 포매팅에서 퍼센트 기호 자체는 %%로 표현합니다.
a = "Error is %d%%"
a
```

```python
# 학습 포인트: 여러 값을 넣을 때는 값들을 괄호로 묶고 순서대로 전달합니다.
a = "i ate %d apples. so i was sick for %s days" % (10,"three")
a
```

```python
# 학습 포인트: 출력 폭과 소수점 자릿수를 지정해 문자열을 정렬할 수 있습니다.
# hi 문자열 출력 시 10개인 공간 사용 
# 오른쪽 정렬 

print("%10s" % "hi")

# 왼쪽 정렬
print("%-10sjane" % "hi")

# 0 : 길이에 상관하지 않음
# 소수점 네 번째 자리까지만
print("%0.4f" % 3.42134234)

# 전체 길이 10개인 문자열 공간에서 오른쪽으로 정렬
print("%10.4f" % 3.42134234)
```

```python
# 학습 포인트: format()의 중괄호 {} 위치에 값이 들어갑니다.
a = "i eat {} apples".format(5)
a
```

```python
# 학습 포인트: 중괄호가 여러 개면 format()의 값도 왼쪽부터 차례대로 들어갑니다.
a = "i ate {} apples. so i was sick for {} days".format(10,"three")
a
```

```python
# 학습 포인트: 같은 결과를 만드는 %, format(), f-string 방식을 비교합니다.
name = "홍길동"
age = 30    
# 첫번째 방법
print("나의 이름은 %s 이고, 나이는 %d 입니다" % (name, age))
# 두번째 방법
print("나의 이름은 {} 이고, 나이는 {} 입니다" .format(name, age))
# f문자열 포매팅
print(f"나의 이름은 {name} 이고, 나이는 {age} 입니다")
```

### 6. 집합 자료형

- 집합에 관련된 것을 쉽게 처리하기 위한 자료형
- 중복 제거 / 순서가 없음
- set()

```python
# 학습 포인트: 세트(set)는 중복을 허용하지 않는 자료형입니다.
s1 = set()
s2 = set([1,2,3,4,5,4])
s3 = set([1,2,3,4,5,6])
```

```python
# 학습 포인트: 중복 값이 제거된 세트의 실제 내용을 출력해 확인합니다.
print(s1)
print(s2)
print(s3)
```

```python
# 학습 포인트: 세트에는 인덱스가 없으므로 순서가 필요하면 리스트나 튜플로 변환합니다.
# TypeError: 'set' object is not subscriptable
# s2[0]

# 인덱싱을, 슬라이싱을 하고 싶다면 가능한 타입으로 변환
# 리스트 : list(), 튜플 : tuple() 

s1 = "안녕하세요"
print(type(s1)) # str
print(type(3)) # int
print(type(list3)) # list
print(type(t4)) # tuple
print(type(dict2)) # dict
print(type(s2)) # set

# set 자료형 => 새로운 list 로 생성
print(list(s2)[2])
print(tuple(s3))
print(s2)
```

```python
# 학습 포인트: set()으로 리스트·문자열을 변환하면 중복이 제거됩니다.
# set() : list,tuple set 구조로 변환시 사용

print(set(list4))
print(set(s1))
```

```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])

# 교집합
print(s1 & s2)
print(s1.intersection(s2))

# 합집합
print(s1 | s2)
print(s1.union(s2))

#차집합
print(s1 - s2)
print(s1.difference(s2))
```

```python
# https://naver.com

# 비밀번호 생성
# 규칙 1. https:// 제외 => naver.com
# 규칙 2. 처음 만나는 . 이후 부분은 제외 => naver
# 규칙 3. 남은 글자 중 처음 세자리 + 글자 갯수 + 글자 내 'e'개수 + '!' 비밀번호 생성 =>nav51!

# 문자 함수, 슬라이싱, str()
```

```python
url = "https://naver.com"

word = url.replace("https://", "")  # naver.com
word = word[:word.find(".")]        # naver

password = word[:3] + str(len(word)) + str(word.count('e')) + '!'

print(password)  # nav51!
```
