#### 인덱싱과 슬라이싱
- 인덱스는 0번 부터 시작

```python
a = "Life is too short, you need ptthon"
print(a[33])
print(a[-1])
```

```python
# 슬라이싱 시 끝번호는 포함하지 않음

print(a[0:5])
print(a[0:2])
print(a[5:7])
print(a[12:17])
print(a[:])
print(a[19:-7])


```

```python
a = "20260714Rainy"
# 날짜와 날씨 분리
print(a[:8])
print(a[8:])

# 년, 월, 일
print(a[:4])
print(a[4:6])


```

#### 문자열 포매팅
- 문자열 안에 어떤 값을 삽입하는 방법

ㅣ 코드 ㅣ 설명 ㅣ
ㅣ --- ㅣ --- ㅣ
ㅣ`%s`ㅣ 문자열(string)  ㅣ
ㅣ`%d ㅣ정수 (integer)   ㅣ
ㅣ %c ㅣ 문자 1개(character) ㅣ
ㅣ %f ㅣ 부동 소수 (floating-point) ㅣ
ㅣ %o ㅣ 8진수 ㅣ
ㅣ %x ㅣ 16진 ㅣ|
%%| % 문자 자체|

- format() 함수 사용


```python
# %s => 어떤 값이든 대입 가능
a = "i eat %d apples" % 10
```

```python
a = "Error is %d%%"
a
```

```python
a = "i ate %d apples. so i was sick for %s days" % (10,"three")
a
```

```python
# hi 문자열 출력 시 10개인 공간 사용 
# 오른쪽 정렬 

print("%10s" % "hi")

# 왼쪽 정렬
print("%-10sjane" % "hi")

# 0 : 길이에 상관하지 않음
# 소수점 네 번째 자리까지만
print("%0.4f" % 3.42134234)

# 전체 길이 10개인 문자열 공간에서 오른쪽으로 정렬
print("%10.4f" % 3.42134234)
```

```python
a = "i eat {} apples".format(5)
a
```
