#### file 읽고 쓰기

| 파일 열기 모드 | 설명 |
|--------------|------|
|`r`|**읽기(Read) 모드** : 파일의 내용을 읽기만 할때 사용. 파일이 존재하지 않는 경우 오류 발생. |
|`w`|**쓰기(Write) 모드** : 파일의 내용을 새로 쓸때 사용. 파일이 존재하지 않는 경우 내용 삭제 후 새로 작성. 없는 경우 새로. |
|`a`
|**추가(Append) 모드** : 기존 파일의 마지막 내용을 추가할때 사용. 파일이 없는 경우 새로 생성.|

```python
# fruits.txt 읽어서 단어의 개수 출
# apple 3
# bannana 2



# apple bannana orange
# 단어 분리 : split() = > ["apple","banana","orange"]

# dict 추가
# key  value
# dict["apple"] += 1
with open("./data/fruits.txt","r",encoding="utf-8")as f:
   contents = f.read().split()
 
dict1 = {}
for apple in contents:
    if apple in dict1: 
        dict1[apple] += 1
    else:
        dict1[apple] = 1

print(dict1)

#보기 좋은 방법
for fruit,count in dict1.items():
    print(f"{fruit} : {count}")


```

```python
# 1000 명의 이름,키,몸무게 info.txt 읽은 후
# BMI 지수 계산
# BMI = 체중 / 키의 제곱
# BMI 가 18.5 미만일 때 저체중
# BMI 가 18.5 ~ 22.9 정상체중
# BMI 가 23 ~ 24.9 과체중
# BMI 가 25 이상 비만

# 이름 : 하타
# 몸무게 : 95
# 키 : 180
# BMI : 18.6
# 결과 : 정상체중

with open("./data/test.txt", "r", encoding="utf-8") as f:

    for line in f:    
        # print(line)
        name,weight,height=f.readline().strip().split(", ")
        # print(name, weight, height)
        weight = int(weight)
        height = int(height)

        bmi = weight / (height / 100) ** 2

        if bmi < 18.5:
            result = "저체중"
        elif bmi < 23:
            result = "정상체중"
        elif bmi < 25:
            result = "과체중"
        else: 
            result = "비만"


        print("\n".join(["이름 : {}","몸무게 : {}","키 : {}","BMI : {:.1f}","결과 : {}"]).format(name,weight,height,bmi,result))
        print()
    
   
```

```python
# origin.txt 읽어서 암호화 시킨 후 encr.txt 저장
# 1. 암호화 선택 : origin.txt 읽어서 변수에 담기 => 그걸 chr(ord("기")+100) 저장
# 2. 복호화 선택 : 암호화된 파일을 읽은 후 원래대로 출력






while True:
    no =int (input("1.암호화 |2.복호화|3.종료 중 원하는 메뉴의 번호를 입력하세요"))
    
    if no == 1:
        with open("./data/origin.txt", "r", encoding="utf-8") as f:
             contents = f.read()
        encrypted = ""

        for char in contents:
            encrypted += (chr(ord(char)+100))
        with open("./data/encr.txt", "w", encoding="utf-8") as f:
            f.write(encrypted)
        print("암호화가 완료되었습니다.")

        
    elif no == 2:
        with open("./data/encr.txt", "r", encoding="utf-8") as f:
            encrypted = f.read()
        decrypted = ""
        for char in encrypted:
            decrypted += (chr(ord(char)-100))

        print(decrypted)
    else:
        break
```
