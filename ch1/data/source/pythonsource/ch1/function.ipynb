````python
# add
# add(2,3)
# add(2,3,4,5)
# add(2,3,4,5,6,7)

# 가변 매개변수
# 입력값이 몇 개가 될 지 모르는 경우
# *
# def 함수명(*변수명):
#     해야할 일


def add(*args):
    result = 0
    for i in args:
        result = result + i
    print(result)
````

````python
# def add_mul(choice, *args, a):
#     pass

# 일반 매개변수는 앞 쪽으로 선언
# +, * 실행
def add_mul(choice, *args):
    if choice == "add":
        result = 0
        for i in args:
            result = result + i

    elif choice == "mul":
        result = 1
        for i in args:
            result = result * i
    
    return result
````

````python
# add_mul("add",10,11,12,13,a=10)
print(add_mul("add",10,11,12,13))
print(add_mul("mul",10,11,12,13))
````

````python
# 키워드 매개변수 (key=value)
# **변수명

# 딕셔너리 == 키워드 매개변수
def print_kwargs(**kwargs):
    print(kwargs)

    # key값만 출력
    for k in kwargs.keys():
        print(k)

    # value 값만 출력
    for v in kwargs.values():
        print(v)

    # key,value 같이 출력
    for k,v in kwargs.items():
        print(k,v)
````

````python
print_kwargs(name="kim",age=15)
print_kwargs(name="park",age=25, addr="서울")
print_kwargs(name="hong",age=30, addr="경기", tel="010-1234-5678")
````

````python
# 변수의 효력 범위

# 함수 밖의 변수 : 전역변수
a = 1
def vartest(a):
    # 매개변수 a : 지역변수
    a = a + 1

vartest(a)
print(a)
````

````python
# 전역변수 a 값 변경하기
# 1) 리턴 이용(추천방법)
a = 1
def vartest(a):
    # 매개변수 a : 지역변수
    a = a + 1
    return a

a = vartest(a)
print(a)
````

````python
# 2) global 키워드 사용
# 함수 안에서 전역변수 접근 안됨

a = 1
def vartest():  
    global a 
    a = a + 1

vartest()
print(a)
````

````python
# 평균 구하는 함수
# get_average()
# 요소의 개수는 정해지지 않음
# 리턴 없음
def get_average(*numbers):
    result = 0
    for i in numbers:
        result = result + i

    print(result / len(numbers))

get_average(10,11,12)
get_average(10,11,12,15,25)
get_average(10,11,12,15,25,36)
````

````python
# 사칙연산 함수
# 결과 리턴

def calculator(num1, num2, op):
    # if~elif
    if op == "+":
        return num1 + num2
    elif op == "-":
        return num1 - num2
    elif op == "*":
        return num1 * num2
    elif op == "/":
        return num1 / num2
    elif op == "//":
        return num1 // num2


print(calculator(10,20,"+"))
print(calculator(10,20,"-"))
print(calculator(10,20,"*"))
print(calculator(10,20,"/"))
print(calculator(10,20,"//"))
````

````python
# filter()
list1 = list(range(1,21))
list1

# 짝수만 새로운 리스트로 생성
# 방법1
even_list = []
for i in list1:
    if i % 2 == 0:
        even_list.append(i)

# 방법2 (리스트 컴프리헨션)
even_list = [i for i in list1 if i % 2 == 0]

# 방법3 filter
# filter(함수, iterable)
list(filter(lambda i:i%2==0, list1))
````

````python
strings = ["Bob","charles","Alexander3","teddy","apple","Banana"]

# 단어의 시작이 대문자인 단어들만 모아서 새로운 리스트로 생성
upper_list = []
for x in strings:
    if x[0].isupper():
        upper_list.append(x)

print(upper_list)

# 리스트 컴프리헨션
upper_list = [x for x in strings if x[0].isupper()]

# filter(함수,원본리스트) : 새로운 리스트로 생성
list(filter(lambda x:x[0].isupper(), strings))
````
