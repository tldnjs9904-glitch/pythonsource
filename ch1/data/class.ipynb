```python
class Student:
    def __init__(self,name,number,grade,detail):
        self.name = name        
        self.number = number        
        self.grade = grade       
        self.detail = detail       

    # 인스턴스 출력할 때 출력시 원하는 값들을 반환
    def __str__(self):
        return f"이름 : {self.name}, 반 : {self.number}, 학번 : {self.grade}, 상세정보 : {self.detail}"
     
```

```python
# 객체 생성 => __init__() 호출
s1 = Student("kim",1,1,{'gender': 'male', 'score1': 95, 'score2': 88})
s2 = Student("Pack",2,2,{'gender': 'female', 'score1': 75, 'score2': 68})
s3 = Student("Hong",2,1,{'gender': 'female', 'score1': 99, 'score2': 86})


# s1은 Student 클래스의 인스턴트임
print(s1)
print(s2)
print(s3)
```
